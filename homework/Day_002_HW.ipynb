## [作業目標]
了解機器學習適合應用的領域與範疇。

## [作業重點]
透過瞭解機器學習的運作方式，找出非常適用機器學習應用的領域及其原因

## 作業
請找出一個你認為機器學習可以發光發熱的領域並闡述原因? 

答：當機器學習與航空業結合在一起，能改善現今整個航空業的生態，而在這之中可以分成幾個領域來討論。
1. 航空器與駕駛員之間的關係：在收集數據的技術完善的現在，航空器製造商在航空器裝載了數以萬計個感測器（sensor），收集飛行及性能數據。而機器學習則可利用這些數據做不同方面的模型訓練，以預測不同的飛行任務及緊急事件。例如：航空器在準備降落之時，若即時感測到跑道長度與當下天氣狀況無法進行安全降落，則會發出警訊，提醒駕駛員應該做出其他的降落方案。
2. 在飛航管制上的機會：飛行員和航管員需要做大量的溝通才能完成空中交通管制工作，儘管主要語言是透過英文，但夾雜著不同風格的口音，將導致難以理解而影響管制的判斷，因此若能透過機器學習，針對飛航管制的語音識別做出訓練，增加飛管兩方的溝通，降低因為溝通上的偏差而產生的飛航事故。
3. 航空公司之內部營運規劃：航班的規劃，一直以來都是航空公司的價值所在。但航班的規劃除了行政上的航權、時間帶分配之外，亦有相當多的內部其他變數存在，包括機組人員的變量（例如休息時間的分配）、飛機燃料及操作限制、飛機座位數量的需求等。因此若能搭配機器學習，便能將複雜的決策規劃過程，降低所需花費的時間，以提供更為彈性的客戶需求。
4. 航空公司之外部行銷規劃：利用機器學習，分析旅客的需求數據，例如：歷史的購買、偏好、目的地、票價、活動等。預測出該提供何種容易吸引顧客的產品（升級座位等手段），吸引、改善客戶航程中的體驗，提高產品的銷售。
